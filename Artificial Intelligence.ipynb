#### Sklearn load, train-test split and standarization example

```python
# import iris form sklearn

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['Species'] = iris.target
iris_df
```

Encode Y

```python
iris_df = pd.get_dummies(iris_df,prefix=['Species'], columns = ['Species'])
iris_df
```

```python
y_len = len(iris.target_names)
X = iris_df.iloc[:,:-y_len]
Y = iris_df.iloc[:,y_len+1:]
X_train, X_test, y_train, y_test = train_test_split(iris_df.iloc[:,:-y_len], iris_df.iloc[:,y_len+1:], test_size=0.2, random_state=42)
X_train
```

```python
y_train
```

```python
sc = StandardScaler().fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

X_train_std
```

#### Visualization

```python
# The indices of the features that we are plotting
x_index = 0
y_index = 1

# this formatter will label the colorbar with the correct target names. 
# Extracted from https://scipy-lectures.org/packages/scikit-learn/auto_examples/plot_iris_scatter.html
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

def plot_iris(df, x, y, formatter):
    plt.figure(figsize=(7, 5))
    plt.scatter(df.data[:, x], df.data[:, y], c=df.target)
    plt.colorbar(ticks=[0, 1, 2], format=formatter)
    plt.xlabel(iris.feature_names[x])
    plt.ylabel(iris.feature_names[y])

    plt.tight_layout()
    plt.show()


plot_iris(iris, 2, 1, formatter)
```

```python
plot_iris(iris, 2, 3, formatter)
```

Deep learning solution

```python
features = X.shape[1]
targets_classes = Y.shape[1]

# Create model
model = Sequential(name='name')
model.add(Dense(50, input_dim=features, activation='relu'))
model.add(Dense(targets_classes, activation='softmax'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Summary
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
# Fit
model.fit(X_train, y_train, batch_size=5, epochs=50, verbose=1, validation_data=(X_test, y_test))
```

```python
preds = np.round(model.predict(X_test))
preds
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
